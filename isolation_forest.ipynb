```python
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# seperate data into training and test set
X = df.iloc[:,:-1]
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# this function creates and returns a model for isolation forest for a given conatmination value
def iso_forest(X_train, cont):
    model = IsolationForest(contamination=cont)
    model.fit(X_train)
    return model
```

```python
# create models with different contamination values
cont_values = [0.001, 0.01, 0.1]
models = [iso_forest(X_train, c) for c in cont_values]

# Make predictions on the test set
y_pred_test = [model.predict(X_test) for model in models]
```

```python
y_pred_test
```

```python
# print number of anomalies predicted for each contamination value
for i in range(3):
    print(f'Number of anomalies for contamination value {cont_values[i]}: {(y_pred_test[i] == -1).sum()}')
```

```python
# modify predicted values to have same class values as original data
modified_pred = [pd.Series(pred).replace({-1: 1, 1: 0}) for pred in y_pred_test]
df_combined = pd.concat(modified_pred, axis=1)
df_combined
```

```python
def plot_figures(index):
    fig, (ax0, ax1) = plt.subplots(1, 2, sharey=True, figsize=(20, 8))

    #original
    ax0.set_title(f'Original (Contamination : {cont_values[index]})')
    ax0.scatter(X_test['V1'], X_test['V2'],c=y_test,cmap='rainbow', s=50)

    # predicted
    ax1.set_title(f'Predicted (Contamination : {cont_values[index]})')
    ax1.scatter(X_test['V1'], X_test['V2'],c=df_combined[index],cmap='rainbow', s=50)
```

```python
# plot graph for attributes V2 and V3 for original and predicted values with contamination value = 0.001
plot_figures(0)
```

```python
# plot graph for attributes V2 and V3 for original and predicted values with contamination value = 0.01
plot_figures(1)
```

```python
# plot graph for attributes V2 and V3 for original and predicted values with contamination value = 0.1
plot_figures(2)
```

```python
# print performance for different hyper-paramaters
def print_performance():
    for i in range(3):
        precision = precision_score(y_test, modified_pred[i])
        recall = recall_score(y_test, modified_pred[i])
        f1 = f1_score(y_test, modified_pred[i])

        print(f"Precision (Contamination : {cont_values[i]}) : {precision:.2f}")
        print(f"Recall    (Contamination : {cont_values[i]}) : {recall:.2f}")
        print(f"F1-score  (Contamination : {cont_values[i]}) : {f1:.2f}")
        print()
```

```python
print_performance()
```
