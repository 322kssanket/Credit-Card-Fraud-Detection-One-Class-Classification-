```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# this function creates and returns a model for DBSCAN for a given epsilon value and min-points
def dbscan(X, epsi, min_pts):
    model = DBSCAN(eps=epsi, min_samples=min_pts)
    lb = model.fit(X)
    return lb.labels_
```

```python
# create model with  hyperparameter values
eps_value = 3 
min_points = 40
X = df.iloc[:, :-1]
y = df['Class']
labels = dbscan(X, eps_value, min_points)
```

```python
# find unique labels
np.unique(labels)
```

```python
# make new column in df
df['dbscan_prediction'] = labels
df['dbscan_prediction'].value_counts()
```

```python
# plot obtained result
plt.figure(figsize=(10,10))
plt.scatter(df['V2'],df['V3'],c=df['dbscan_prediction'],cmap='rainbow',s=15)
plt.title(f'DBSCAN Clustering (Epsilon : {eps_value}, Min-Points : {min_points})')
plt.xlabel('V2')
plt.ylabel('V3')
plt.show()
```

```python
# print performance
def print_performance():
    recall = recall_score(y, labels, average='weighted', zero_division=1)
    f1 = f1_score(y, labels, average='weighted', zero_division=1)

    print(f"Recall   : {recall:.2f}")
    print(f"F1-score : {f1:.2f}")
    print()
```

```python
print_performance()
```
